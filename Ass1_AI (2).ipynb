```python
graph = {
 'A' : ['B','C'],
 'B' : ['D', 'E'],
 'C' : ['F'],
 'D' : [],
 'E' : ['F'],
 'F' : []
} #dictionary
visited = [] #List
queue = [] # list
#how to create variables in python
def bfs(visited, graph, node): #node => Starting node of graph
    visited.append(node)
    queue.append(node)
    while queue:
        s = queue.pop(0)
        print (s, end = " ")
        for neighbour in graph[s]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

bfs(visited, graph, 'A') #function calling

#The time complexity of BFS is O(V+E)
#The space complexity of BFS is O(V)
```

```python

graph = {
'A' : ['B','C'],
'B' : ['D', 'E'],
'C' : ['F'],
'D' : [],
'E' : ['F'],
'F' : []
}
visited = set()
def dfs(visited, graph, node): #function for dfs
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)  #recursive algorithm

print("Following is the Path using Depth-First Search")
dfs(visited, graph, 'A')

#The time complexity of BFS is O(V+E)
#The space complexity of BFS is O(V)
```

```python
def create_graph():
    graph = {}
    while True:
        node = input("Enter node (or 'exit' to stop): ")
        if node == 'exit':
            break
        neighbors = input(f"Enter neighbors for node {node} separated by spaces: ").split()
        graph[node] = neighbors
    return graph


user_graph = create_graph()
print("User-generated graph:", user_graph)
```
