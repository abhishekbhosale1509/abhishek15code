```python
def solve_queens(board=None, row=0):
    if board is None:
        board = [[0] * 8 for _ in range(8)]

    if row >= len(board): # if queen has been sucessfully placed then it return true.
        print("Solution found:")
        for row in board:
            print(row)
        return True

    for col in range(len(board)):
        if all(board[i][col] == 0 for i in range(row)) and \
           all(board[i][j] == 0 for i, j in zip(range(row, -1, -1), range(col, -1, -1))) and \
           all(board[i][j] == 0 for i, j in zip(range(row, -1, -1), range(col, len(board)))):
            board[row][col] = 1
            if solve_queens(board, row + 1): # If a solution is found, it returns True. If not, then backtracks put 0 back.
                return True
            board[row][col] = 0
    return False     # if current configuration doesn't lead to a solution then return false.

# 1] check there is no queen in the same column
# 2] check no queen in the upper left diagonal
# 3] check no queen in the upper right diagonal

# recursively calls solve_queens with the next row, and checks if a solution can be found.
# If a solution is found, it returns True. If not,
# it backtracks by removing the queen (setting the position back to 0) and tries the next column.

solve_queens()


# The all() function returns True if all items in an iterable are true, otherwise it returns False.
# Python's zip() function creates an iterator that will aggregate elements from two or more iterables.
# Time Complexity := O(N^2 + N^N)
# Space Complexity := O(N^2)
```
